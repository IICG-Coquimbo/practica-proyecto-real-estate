```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("EDA_RealEstate_Melany") \
    .config(
        "spark.jars.packages",
        "org.mongodb.spark:mongo-spark-connector_2.12:10.1.1"
    ) \
    .config(
        "spark.mongodb.read.connection.uri",
        URI_MONGO
    ) \
    .config(
        "spark.mongodb.write.connection.uri",
        URI_MONGO
    ) \
    .config(
        "spark.mongodb.read.database",
        "BD_RealEstate"
    ) \
    .config(
        "spark.mongodb.read.collection",
        "Propiedades"
    ) \
    .getOrCreate()

print("Spark Atlas conectado")
```

```python
df = spark.read.format("mongodb").load()
```

```python
print(df.count())
```

```python
df.printSchema()
```

```python
df.show(5, truncate=False)
```

```python
from pyspark.sql.functions import col, count, when

df.select([
    count(when(col(c).isNull(), c)).alias(c)
    for c in df.columns
]).show()
```

```python
df.describe([
    "precio",
    "m2",
    "dormitorios",
    "banos"
]).show()
```

```python
from pyspark.sql import functions as F

df.groupBy("dormitorios") \
    .agg(
        F.avg("precio").alias("precio_promedio")
    ) \
    .orderBy("dormitorios") \
    .show()
```

```python
df.groupBy("ubicacion") \
    .agg(
        F.avg("precio").alias("precio_promedio"),
        F.count("*").alias("cantidad")
    ) \
    .orderBy(F.desc("cantidad")) \
    .show(truncate=False)
```

```python
df.groupBy("piscina") \
    .count() \
    .show()
```

```python
df.select(
    "titulo",
    "ubicacion",
    "precio",
    "m2"
).orderBy(
    F.desc("precio")
).show(10, truncate=False)
```

```python
df_limpio = df.filter(
    (col("precio") > 100000) &
    (col("precio") < 5000000) &
    (col("m2") > 10)
)

print("Registros originales:", df.count())
print("Registros limpios:", df_limpio.count())
```
