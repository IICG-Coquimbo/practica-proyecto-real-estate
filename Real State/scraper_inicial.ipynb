```python
import requests
from bs4 import BeautifulSoup

# 1. Definir la URL (Ellos cambiarán esto según su caso)
url = "https://cl.agentiz.com"

# 2. Hacer la petición
headers = {'User-Agent': 'Mozilla/5.0'} #Para que no nos bloqueen
response = requests.get(url, headers=headers)

# 3. Convertir el HTML en un objeto de Python. 
soup = BeautifulSoup(response.text, 'html.parser')


# 3. Buscar los datos (Ejemplo: buscar todos los titulos)
# Cambia ’h2’ y ’nombre-de-la-clase’ por lo que inspeccionaste
resultados = sopa.find_all(’div’, class_=’nombre-de-la-clase’)

# 4. Mostrar en pantalla
for dato in resultados:
    print(f"Encontrado: {dato.text.strip()}")
```
