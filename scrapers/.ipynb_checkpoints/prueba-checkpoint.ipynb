```python
import os
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from webdriver_manager.core.os_manager import ChromeType

print("1. Configurando opciones para Brave...")
options = Options()
options.binary_location = "/usr/bin/brave-browser"
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

try:
    print("2. Descargando el driver compatible...")
    # Forzamos la descarga del driver específico para Brave
    ruta_driver = ChromeDriverManager(chrome_type=ChromeType.BRAVE).install()
    print(f" -> Driver descargado en: {ruta_driver}")
    
    print("3. Intentando abrir el navegador...")
    driver = webdriver.Chrome(service=Service(ruta_driver), options=options)
    
    print("¡ÉXITO! El navegador se abrió correctamente.")
    driver.get("https://www.google.com")
    print(f"Título de la página: {driver.title}")
    
    driver.quit()
except Exception as e:
    print(f"\nERROR DETALLADO: {e}")
```
