```python
import requests
from bs4 import BeautifulSoup
from pymongo import MongoClient
import time

# =========================
# MONGO
# =========================
client = MongoClient("mongodb://bigdata_mongodb:27017/")
db = client["Inmobiliaria"]
coleccion = db["books"]

# =========================
# URL
# =========================
url = "https://books.toscrape.com/"
headers = {"User-Agent": "Mozilla/5.0"}

respuesta = requests.get(url, headers=headers)
soup = BeautifulSoup(respuesta.text, "html.parser")

# =========================
# SCRAPING
# =========================
productos = soup.find_all("article", class_="product_pod")

print("Total encontrados:", len(productos))

datos = []

for p in productos:
    titulo = p.h3.a["title"]
    precio = p.find("p", class_="price_color").text

    precio_limpio = precio.replace("£", "").replace("Â", "").strip()
    precio_num = float(precio_limpio)

    data = {
        "titulo": titulo,
        "precio": precio_num,
        "precio_texto": precio,
        "fecha": time.strftime("%Y-%m-%d %H:%M:%S"),
        "grupo": "G2_Inmobiliaria"
    }

    datos.append(data)

    print(titulo, "-", precio)

# =========================
# GUARDAR
# =========================
if datos:
    coleccion.insert_many(datos)
    print(f"{len(datos)} guardados en MongoDB")
else:
    print("❌ No se encontraron datos")
```
