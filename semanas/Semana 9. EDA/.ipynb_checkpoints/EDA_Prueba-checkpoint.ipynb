```python
import pymongo
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Tu cadena de conexión real de Atlas
client = pymongo.MongoClient("mongodb+srv://bd_realestate:abc123456@c-realestate.xyfip8o.mongodb.net/?appName=C-RealEstate")
db = client["BD_RealEstate"]
collection = db["Propiedades"]

# Descargar los datos directo a un DataFrame de Pandas
datos = list(collection.find())
df_raw = pd.DataFrame(datos)

print(f"¡Conectado con éxito! Registros reales traídos de Coquimbo: {len(df_raw)}")
```

```python
# ==============================================================================
# CELDA 2 CORREGIDA: LIMPIEZA ADAPTADA A DECIMALES EN STRINGS
# ==============================================================================

# Filtramos los nulos en las columnas base de tus datos reales
df = df_raw.dropna(subset=['precio', 'dormitorios', 'banos']).copy()

# 1. Limpiar precios (quitar signos de dinero, puntos de miles, espacios)
df['precio_num'] = df['precio'].astype(str).str.replace(r'[\$\.\sA-Za-z]', '', regex=True)
df['precio_num'] = pd.to_numeric(df['precio_num'], errors='coerce')

# 2. Limpiar dormitorios (Primero lo pasamos a float para absorber el '.0' y luego a entero)
df['dormitorios_clean'] = df['dormitorios'].astype(str).str.replace(r'[\sA-Za-z]', '', regex=True)
df['dormitorios_num'] = pd.to_numeric(df['dormitorios_clean'], errors='coerce').fillna(0).astype(int)

# 3. Limpiar baños (Mismo proceso seguro)
df['banos_clean'] = df['banos'].astype(str).str.replace(r'[\sA-Za-z]', '', regex=True)
df['banos_num'] = pd.to_numeric(df['banos_clean'], errors='coerce').fillna(0).astype(int)

# 4. Filtrar filas inconsistentes o con ceros
df = df[(df['precio_num'] > 0) & (df['dormitorios_num'] > 0) & (df['banos_num'] > 0)]

# 5. Crear la columna de Ratio Estructural para tu análisis
df['ratio_config'] = df['dormitorios_num'].astype(str) + "D / " + df['banos_num'].astype(str) + "B"

print("¡Datos de Coquimbo limpiados con éxito!")
print(df[['precio_num', 'dormitorios_num', 'banos_num', 'ratio_config']].head())
```

```python
# Gráfico 1: La Guerra de los Baños
plt.figure(figsize=(9, 5))
sns.boxplot(x='banos_num', y='precio_num', data=df, palette="Blues")
plt.title("La Guerra de los Baños en Coquimbo: Impacto del Segundo Baño")
plt.xlabel("Cantidad de Baños")
plt.ylabel("Precio de Arriendo ($)")
plt.grid(True)
plt.show()

# Gráfico 2: Rendimiento Decreciente de los Dormitorios
plt.figure(figsize=(9, 5))
sns.lineplot(x='dormitorios_num', y='precio_num', data=df, marker='o', color='orange', errorbar=None)
plt.title("Curva de Rendimiento Decreciente de Dormitorios")
plt.xlabel("Cantidad de Dormitorios")
plt.ylabel("Precio Promedio ($)")
plt.grid(True)
plt.show()
```
