```python
import pymongo
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# ==========================================
# PASO 1: TRAER LOS DATOS DESDE ATLAS (Para abrir la memoria del nuevo notebook)
# ==========================================
client = pymongo.MongoClient("mongodb+srv://bd_realestate:abc123456@c-realestate.xyfip8o.mongodb.net/?appName=C-RealEstate")
db = client["BD_RealEstate"]
collection = db["Propiedades"]

df_raw = pd.DataFrame(list(collection.find()))

# ==========================================
# PASO 2: LIMPIEZA EXPRESA PARA QUE ENTIENDA LOS DECIMALES ('1.0')
# ==========================================
df = df_raw.dropna(subset=['precio', 'dormitorios', 'banos']).copy()

df['precio_num'] = df['precio'].astype(str).str.replace(r'[\$\.\sA-Za-z]', '', regex=True)
df['precio_num'] = pd.to_numeric(df['precio_num'], errors='coerce')

df['dormitorios_clean'] = df['dormitorios'].astype(str).str.replace(r'[\sA-Za-z]', '', regex=True)
df['dormitorios_num'] = pd.to_numeric(df['dormitorios_clean'], errors='coerce').fillna(0).astype(int)

df['banos_clean'] = df['banos'].astype(str).str.replace(r'[\sA-Za-z]', '', regex=True)
df['banos_num'] = pd.to_numeric(df['banos_clean'], errors='coerce').fillna(0).astype(int)

df = df[(df['precio_num'] > 0) & (df['dormitorios_num'] > 0) & (df['banos_num'] > 0)]

# Guardamos el cluster_id para usarlo en los perfiles
df['ratio_config'] = df['dormitorios_num'].astype(str) + "D / " + df['banos_num'].astype(str) + "B"

print(f"¡Base de datos cargada y reconocida en Semana 10! Registros listos: {len(df)}")

# ==========================================
# PASO 3: TU ALGORITMO DE CLUSTERING (K-MEANS)
# ==========================================
X = df[["dormitorios_num", "banos_num", "precio_num"]]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Buscamos tus 3 perfiles de departamentos en Coquimbo
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['cluster_id'] = kmeans.fit_predict(X_scaled)

score = silhouette_score(X_scaled, df['cluster_id'])
print(f"Calidad del agrupamiento (Silhouette Score): {score:.4f}\n")

print("=== PERFILAMIENTO DE LOS CLUSTERS DESCUBIERTOS EN COQUIMBO ===")
print(df.groupby('cluster_id')[['precio_num', 'dormitorios_num', 'banos_num']].mean())
```

```python
print("=========================================================")
print("            CANTIDAD DE PROPIEDADES POR CLUSTER          ")
print("=========================================================")
conteo_clusters = df['cluster_id'].value_counts()
print(conteo_clusters)
```

```python
import numpy as np
from sklearn.decomposition import PCA

# 1. Aplicamos PCA sobre los datos escalados anteriores
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# 2. Guardamos los componentes en el DataFrame
df['PC1'] = X_pca[:, 0]
df['PC2'] = X_pca[:, 1]

# 3. Graficamos exactamente con el estilo del notebook de la profesora
plt.figure(figsize=(10, 6))
sns.set_style("darkgrid")

# Calculamos límites basados en percentiles para evitar que datos atípicos deformen el gráfico
pc1_min, pc1_max = np.percentile(df['PC1'], [1, 99])
pc2_min, pc2_max = np.percentile(df['PC2'], [1, 99])

# Dibujar el Scatter Plot usando el cluster_id como color (hue)
sns.scatterplot(
    data=df,
    x='PC1',
    y='PC2',
    hue='cluster_id',
    palette='viridis',
    s=70,
    alpha=0.6
)

plt.xlim(pc1_min - 1, pc1_max + 1)
plt.ylim(pc2_min - 1, pc2_max + 1)
plt.title('Visualización de los Clusters de Departamentos en Coquimbo (PCA)')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()
```

```python
from sklearn.cluster import DBSCAN

# Ejecutamos DBSCAN con los parámetros estándar para comparar
dbscan = DBSCAN(eps=0.5, min_samples=5)
df['cluster_dbscan'] = dbscan.fit_predict(X_scaled)

print("=== DISTRIBUCIÓN DE CLUSTERS CON ALGORITMO DBSCAN ===")
print(df['cluster_dbscan'].value_counts())
print("\n*Nota: El valor -1 representa el ruido o propiedades fuera de lo común detectadas en Coquimbo.")
```
