```python
import pymongo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# ==============================================================================
# BLOQUE 1: CONEXIÓN Y CARGA DE DATOS DESDE ATLAS
# ==============================================================================
client = pymongo.MongoClient("mongodb+srv://bd_realestate:abc123456@c-realestate.xyfip8o.mongodb.net/?appName=C-RealEstate")
db = client["BD_RealEstate"]
collection = db["Propiedades"]

df_raw = pd.DataFrame(list(collection.find()))

# Limpieza estricta de textos a números reales (absorbiendo el '.0')
df = df_raw.dropna(subset=['precio', 'dormitorios', 'banos']).copy()
df['precio_num'] = pd.to_numeric(df['precio'].astype(str).str.replace(r'[\$\.\sA-Za-z]', '', regex=True), errors='coerce')
df['dormitorios_num'] = pd.to_numeric(df['dormitorios'].astype(str).str.replace(r'[\sA-Za-z]', '', regex=True), errors='coerce').fillna(0).astype(int)
df['banos_num'] = pd.to_numeric(df['banos'].astype(str).str.replace(r'[\sA-Za-z]', '', regex=True), errors='coerce').fillna(0).astype(int)

# Filtro de outliers para estabilizar la pendiente lineal en Coquimbo
df = df[(df['precio_num'] > 0) & (df['precio_num'] < 5000000) & (df['dormitorios_num'] > 0) & (df['banos_num'] > 0)]

print(f"¡Base de datos cargada en Semana 11! Registros para regresión: {len(df)}")

# ==============================================================================
# BLOQUE 2: SEGREGACIÓN DE VARIABLES Y SPLIT (80/20)
# ==============================================================================
X_reg = df[["dormitorios_num", "banos_num"]]
y_reg = df["precio_num"]

X_train, X_test, y_train, y_test = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# ==============================================================================
# BLOQUE 3: ENTRENAMIENTO DEL MODELO DE REGRESIÓN LINEAL
# ==============================================================================
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Predicciones para testeo
predicciones = lr_model.predict(X_test)

# ==============================================================================
# BLOQUE 4: EVALUACIÓN DE DESEMPEÑO (R² y RMSE - METRICAS DE LA PROFE)
# ==============================================================================
r2 = r2_score(y_test, predicciones)
rmse = np.sqrt(mean_squared_error(y_test, predicciones))

print("\n==================================================")
print("       EVALUACIÓN DEL MODELO DE REGRESIÓN")
print("==================================================")
print(f"Coeficiente de Determinación (R²): {r2 * 100:.2f}%")
print(f"RMSE (Error promedio del modelo):  ${rmse:,.2f} CLP")
print("==================================================")

print("\n==================================================")
print("         INTERPRETACIÓN DE LOS COEFICIENTES")
print("==================================================")
print(f"Intersección (Precio base construido):   ${lr_model.intercept_:,.2f}")
print(f"Coeficiente marginal por Dormitorio:      ${lr_model.coef_[0]:,.2f}")
print(f"Coeficiente marginal por BAÑO adicional:  ${lr_model.coef_[1]:,.2f}")
print("==================================================")

# ==============================================================================
# BLOQUE 5: GRÁFICOS COMPLEMENTARIOS (PREDICCIONES VS REALES Y RESIDUOS)
# ==============================================================================
plt.figure(figsize=(12, 5))

# Gráfico 1: Valores Reales vs Predicciones
plt.subplot(1, 2, 1)
plt.scatter(y_test, predicciones, color='blue', alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.title("Valores Reales vs Predicciones")
plt.xlabel("Precio Real ($)")
plt.ylabel("Precio Predicho ($)")

# Gráfico 2: Análisis de Residuos (Errores)
residuos = y_test - predicciones
plt.subplot(1, 2, 2)
sns.histplot(residuos, kde=True, color='purple')
plt.title("Distribución de los Residuos (Errores)")
plt.xlabel("Error de Predicción ($)")

plt.tight_layout()
plt.show()
```
